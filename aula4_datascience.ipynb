```python
#importando as bibliotecas que contem comandos 
import pandas as pd
import numpy as np
import io
import folium
from google.colab import files
import plotly.express as px
```

```python
#carregando um arquivo do meu computador 
file_uploaded = files.upload()
```

```python
#Carrega o arquivo da nuvem para a IDE do google colabs
df = pd.read_csv (io.BytesIO (file_uploaded['AB_NYC_2019.csv'] ) )
```

```python
#removendo as linhas faltantes
df1 = df.dropna()
```

```python
#Removendo linhas com nenhuma disponibilidade de locação
df1.dropna()
df1 = df1.loc[df1['availability_365']>0,:]
```

```python
#1.Qual o valor do aluguel (diária) mais caro de cada região da base de dados da cidade de NY, apenas para os imóveis disponíveis para alugar
#A diária mais cara da região do bronx é U$ 800
#A diária mais cara da região do Brooklyn é U$ 8000
#A diária mais cara da região de Manhatam é U$ 10.000
#A diária mais cara da região do Queens U$ 2600
#A diária mais cara da região de Staten Island é U$ 625
cols =['neighbourhood_group','price']
data_plot = df1.loc[:,cols].groupby('neighbourhood_group').max().reset_index()

px.bar(data_plot,x= 'neighbourhood_group',y='price')
```

```python
#Conseguimos saber onde estão localizados os imóveis com o valor do aluguel mais caro, na cidade de NY

#Conjunto de dados com as coordenadas
cols= ['price','neighbourhood_group','latitude','longitude']
data_plot = df1.loc[:,cols].groupby('neighbourhood_group').max().reset_index()

#Um comando para o mapa
map = folium.Map()

#Colocar os pontos de longitude e latitude dentro do mapa
for index, location_info in data_plot.iterrows():
 folium.Marker( [location_info['latitude'],location_info['longitude']],popup=location_info['neighbourhood_group']).add_to(map)


map
```

```python
#Conseguimos saber onde estão localizados os imóveis pelo seu tipo?
#Conjunto de dados
cols = ['neighbourhood_group','room_type','latitude', 'longitude']
data_plot=df1.loc[:,cols].sample(100)

#Criar uma nova coluna
data_plot.loc[:,'color'] = 'NA'

data_plot.loc[data_plot['room_type'] == 'Private room', 'color'] = 'darkgreen'
data_plot.loc[data_plot['room_type'] == 'Entire home/apt','color'] = 'darkred'
data_plot.loc[data_plot['room_type'] == 'Shared room', 'color'] = 'purple'

#Desenhar mapa 
map = folium.Map ()

#Colorcar os pontos dentro do mapa
#Colocar os pontos de latitude e longitude dentro do mapa

for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'],location_info['longitude']],
                popup=location_info['neighbourhood_group'], 
                icon=folium.Icon( color=location_info['color']) ).add_to(map)

map
```
